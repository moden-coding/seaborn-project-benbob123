# Part 4: Data Visualization

## Plot distribution of individual variables
Use histograms, KDE plots, box plots, and violin plots (choose 1 - 2) to show the distribution of data within 3 selected numerical columns. Be sure to include histograms to identify outliers

```python
pop_correlation_matrix = population.corr()['Uninsured Rate (2015)'].sort_values().to_frame()
sns.heatmap(pop_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

This heatmap shows how correlated different demographic data are to the uninsured rate.
From this visulaization we can noice that 'Female Persons' is the demogrphic that has the strongest correlation to uninsured rate. Keep in mind that this is a negative correlation


```python
sns.regplot(x=population['Female persons, percent, 2014'], y=population['Uninsured Rate (2015)'], scatter_kws={'s': 30}, line_kws={'color': 'red'})

# Add labels and title
plt.xlabel('Female persons, percent, 2014 (%)')
plt.ylabel('Uninsured Rate (2015) (%)')
plt.title('')

# Show the plot
plt.show()
```

From this regplot we can see the negative correlation between the percentage of females in each state and the state's uninsured rate. From the heatmap we learned that the r value is only -0.36 so as we can see here, the correlation esists; it just isn't strong

```python
politicalCor = politicsDF[['Conservative','Moderate','Liberal']]
politicalCor['Uninsured Rate (2015)'] = merged_df['Uninsured Rate (2015)']

political_correlation_matrix = politicalCor.corr()['Uninsured Rate (2015)'].sort_values().to_frame()
sns.heatmap(political_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

From this heatmap we learn that Liberals are more likely to have health insurance while Conservatives are less likely to have health insurance

```python
\

# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plot the first regression plot on the first subplot
sns.regplot(x='Conservative', y='Uninsured Rate (2015)', data=politicalCor, label='Conservative (%)', ax=axes[0],color='red')

# Plot the second regression plot on the second subplot
sns.regplot(x='Liberal', y='Uninsured Rate (2015)', data=politicalCor, label='Liberal (%)', ax=axes[1],color='blue')

# Show legend on both subplots
axes[0].legend()
axes[1].legend()

# Set title for each subplot
axes[0].set_title('Conservative')
axes[1].set_title('Liberal')

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Display the plot
plt.show()
```

These regplots are just another way to illustrate that Conservatives are more likely to be uninsured while Liberals are less likely to be uninsured.
